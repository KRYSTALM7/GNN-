```python
import random
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch_geometric.nn import GCNConv
from torch_geometric.data import Data, DataLoader
import matplotlib.pyplot as plt
import networkx as nx


# Generate the data
w = [random.randint(1, 10) for _ in range(100)]
x = [random.randint(11, 20) for _ in range(100)]
y = [random.randint(21, 30) for _ in range(100)]
z = [random.randint(31, 40) for _ in range(100)]
df = pd.DataFrame({'time': range(100)})
df['w'] = w
df['x'] = x
df['y'] = y
df['z'] = z
df.index = df['time']
df = df.drop(columns=['time'])

w_test = [random.randint(1, 10) for _ in range(50)]
x_test = [random.randint(11, 20) for _ in range(50)]
y_test = [random.randint(21, 30) for _ in range(50)]
z_test = [random.randint(31, 40) for _ in range(50)]
df_test = pd.DataFrame({'time': range(50)})
df_test['w'] = w_test
df_test['x'] = x_test
df_test['y'] = y_test
df_test['z'] = z_test

df_test['w'][10:12] = 70
df_test['x'][10:12] = 98
df_test['y'][10:12] = -30
df_test['z'][10:12] = 300

df_test['w'][30:32] = 70
df_test['x'][30:32] = 98
df_test['y'][30:32] = -30
df_test['z'][30:32] = 300

df_test.index = df_test['time']
df_test = df_test.drop(columns=['time'])

# Convert the pandas dataframes to PyTorch tensors
train_data = torch.tensor(df.values, dtype=torch.float32)
test_data = torch.tensor(df_test.values, dtype=torch.float32)

# Define the GNN model
class GNN(nn.Module):
    def __init__(self, input_dim, hidden_dim):
        super(GNN, self).__init__()
        self.conv1 = GCNConv(input_dim, hidden_dim)
        self.conv2 = GCNConv(hidden_dim, input_dim)

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = self.conv2(x, edge_index)
        return x

# Create a PyTorch geometric Data object for the training data
train_edges = torch.tensor([[i, i + 1] for i in range(len(train_data) - 1)], dtype=torch.long).t().contiguous()
train_graph = Data(x=train_data, edge_index=train_edges)

# Create a PyTorch geometric DataLoader for batch training
train_loader = DataLoader([train_graph], batch_size=1, shuffle=True)

# Initialize the GNN model and define the optimizer
model = GNN(input_dim=len(df.columns), hidden_dim=64)
optimizer = optim.Adam(model.parameters(), lr=0.01)

# Train the GNN model
model.train()
for epoch in range(100):
    for batch in train_loader:
        optimizer.zero_grad()
        out = model(batch.x, batch.edge_index)
        loss = F.mse_loss(out[:-1], batch.x[1:])  # Mean Squared Error loss
        loss.backward()
        optimizer.step()

# Test the GNN model on the test data
model.eval()
with torch.no_grad():
    test_edges = torch.tensor([[i, i + 1] for i in range(len(test_data) - 1)], dtype=torch.long).t().contiguous()
    test_graph = Data(x=test_data, edge_index=test_edges)
    test_output = model(test_graph.x, test_graph.edge_index)
    
# Compute the anomaly scores for each variable
anomaly_threshold = 50.0  # Adjust this threshold as needed
anomaly_mask = torch.abs(test_data - test_output) > anomaly_threshold
anomaly_scores = torch.mean(torch.abs(test_data - test_output), dim=1)

# Plot the results
plt.figure(figsize=(12, 8))
for i in range(len(df.columns)):
    plt.subplot(len(df.columns), 1, i+1)
    plt.plot(test_data[:, i], label='Original Data', color='blue')
    plt.plot(test_output[:, i], label='Predicted Data', color='green')
    plt.scatter(torch.nonzero(anomaly_mask[:, i]).squeeze().tolist(), test_output[:, i][anomaly_mask[:, i]].squeeze().tolist(),
                color='red', label='Anomalies')
    plt.xlabel('Time')
    plt.ylabel('Value')
    plt.legend()
plt.suptitle('GNN Anomaly Detection')
plt.tight_layout()
plt.show()

# Print the anomaly detection results
if torch.any(anomaly_mask):
    print("Anomalies detected!")
    print("Anomaly indices:")
    for i in range(len(df.columns)):
        print(f"Variable {df.columns[i]}:", torch.nonzero(anomaly_mask[:, i]).squeeze().tolist())
else:
    print("No anomalies detected.")
```
